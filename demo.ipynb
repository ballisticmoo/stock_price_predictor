```python
import Stonks
import pandas as pd
```

```python
df = pd.read_csv("datasets/BPHLY.csv", parse_dates=['Date'])
df.dtypes
```

```python
df
```

```python
data, scaler = Stonks.Processing.preprocess_stock_data(df, future_steps=7)
data
```

```python
model = Stonks.Model.load('lstm')
```

```python
# Assuming you already have a preprocessed DataFrame called 'preprocessed_data'
independent = ['Close_future_1','Close_future_2', 'Close_future_3', 'Close_future_4', 'Close_future_5','Close_future_6', 'Close_future_7']

X = data.drop(['Close', 'Date'] + independent, axis=1).values
y = data[independent].values
```

```python
y_pred = model.predict(X)
```

```python
scaler.data_min_
```

```python
data.drop(['Date',], axis=1).iloc[45]
```

```python
inversed = scaler.inverse_transform(data.drop(['Date',], axis=1))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
start_index = 50
plot_x_lookback = inversed[start_index][0:30].round(2)[::-1]
plot_x_lookback
```

```python
inversed_y_pred = []
for i in y_pred[start_index]:
    reversed = round((i * (scaler.data_max_[31] - scaler.data_min_[31]) + scaler.data_min_[31]), 2)
    inversed_y_pred.append(reversed)
```

```python
inversed_y_pred
```

```python
date = range(1,38)
plt.plot(date, np.concatenate((plot_x_lookback, inversed[start_index][31:38]), axis=0 ), label='Actual')
plt.plot(date[30:37], np.array(inversed_y_pred), label='Predictions')

plt.title("Bank of the Philippine Islands (BPHLY)")
plt.xlabel("Days")
plt.ylabel("Stock Price ($)")
plt.legend()
plt.grid(True)


plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(inversed_y_pred, inversed[start_index][31:38])
mae = mean_absolute_error(inversed_y_pred, inversed[start_index][31:38])
print(f"MSE: {mse:,.4f}, MAE: {mae:,.4f}")

usd_to_php = 57.24
print(f"MSE: {mse * usd_to_php:,.4f}, MAE: {mae * usd_to_php:,.4f}")
```
